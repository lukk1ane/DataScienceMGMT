```python
import pandas as pd
from pandas import Series
```

```python
data = [10,20,30,40,50]
labels = ['a','b','c','d','e']
t=pd.Series(data=data,index=labels)
print(t)
```

```python
data = [10,20,30,40,50]
labels = ['a','b','c','d','e']
t=pd.Series(data=data,index=labels)
finder = t.iloc[2]
print(finder)
finders=t.loc['b']
print(finders)
```

```python
data1=[5,10,15,20]
labels1=['w','x','y','z']
t=pd.Series(data=data1,index=labels1)
data2=[2,4,6,8]
labels2=['w','x','y','p']
p=pd.Series(data=data2,index=labels2)
sum = t.add(p,fill_value=0)
print(sum)
sub=t.subtract(p,fill_value=0)
print(sub)
```

```python
data = {
    'Item':[ "Computer", "Fan", "Chip", "Power", "Card"
    ],
    "Cost":[100,200,300,400,500]
}
t=pd.DataFrame(data)
t['Tax']=t['Cost']*0.1
t['Total']=t['Cost'] + t['Tax']
filter  = t[t['Total']>50]
print(filter)
```

```python
data = {
    'Product': ['Laptop', 'Mouse', 'Desk', 'Chair', 'Monitor', 'Keyboard',
                'Lamp', 'Notebook', 'Pen', 'Headphones'],
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture',
                 'Electronics', 'Electronics', 'Furniture', 'Stationery',
                 'Stationery', 'Electronics'],
    'Price': [800, 25, 300, 450, 350, 60, 80, 5, 2, 150],
    'Stock': [15, 50, 8, 12, 20, 35, 25, 100, 200, 30]
}
t=pd.DataFrame(data)
isnull=t.isnull()
count=t.value_counts('Category')
exp=t.nlargest(3,'Price')
cheap=t.nsmallest(2,'Price')
print(t.shape)
print(t.ndim)
```

```python
data = {
    'Student': ['Alice', 'Alice', 'Bob', 'Bob', 'Charlie', 'Charlie',
                'David', 'David', 'Eve', 'Eve', 'Alice', 'Bob'],
    'Subject': ['Math', 'Science', 'Math', 'Science', 'Math', 'Science',
                'Math', 'English', 'Science', 'English', 'English', 'English'],
    'Score': [85, 92, 78, 88, 95, 89, 72, 80, 91, 87, 88, 75],
    'Attendance': [95, 90, 85, 92, 98, 96, 80, 78, 93, 88, 92, 83]
}
t=pd.DataFrame(data)
drop = t.drop_duplicates(subset='Student', keep='first')
index = drop.set_index('Student')

print(index[index['Score'].between(70,90)])
```
