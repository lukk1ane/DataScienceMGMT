```python
def fizz_buzz(n):
    for i in range(1,n+1):
        if i%3==0 and i%5==0:
            print("FizzBuzz")

        elif i%3==0:
            print("Fizz")
        elif i%5==0:
            print("Buzz")
        else:
            print(i)

print(fizz_buzz(20))
```

```python
def division(num1,num2):
    try:
        return num1/num2
    except ZeroDivisionError:
        print("Division by zero")
print(division(3,5))
```

```python
def list_index(list,index):
    try:
        return list[index]
    except IndexError:
        return "out of range"
list=[1,2,3,4,5,6,7,8]
print(list_index(list,19))
```
